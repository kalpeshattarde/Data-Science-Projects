# Graphura Influencer Recommendation System

#### 1. INTRODUCTION

Graphura LinkedIn Content Intelligence is a data-driven system designed to analyze past LinkedIn performance and recommend the most effective content strategies to maximize engagement and visibility.

It combines historical trends, machine learning models, and content metadata to produce actionable insights tailored for LinkedIn content planning.

## 1. Importing essential python libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings('ignore')
```

## 2. Dataset Loading

```python
data = pd.read_excel("Final combined influencer data.xlsx")
data.head()
```

## 3. Data Cleaning & Preprocessing

```python
data.info()
data.isnull().sum()
data.fillna('0',inplace=True)
data.head()
```

```python
data.info()
```

```python
data["Engagement Rate (%)"] = (
    (data["Average Likes"].fillna(0) + data["Average Comments"].fillna(0))
    / data["Follower Count"].replace(0, np.nan)
) * 100
```

```python
def assign_segment(followers):
    if followers < 50000:
        return "Micro"
    elif followers < 250000:
        return "Mid-tier"
    else:
        return "Macro"

data["Segment"] = data["Follower Count"].apply(assign_segment)
```

```python
unc = ['Creators Name','Contact Email']
df = data.drop(columns=unc)

df.head()
```

```python
# Clean text first
df["Past Brand Collaborations"] = (
    df["Past Brand Collaborations"]
    .astype(str)
    .str.strip()
    .str.lower()
)

# Force replace using loc
df.loc[df["Past Brand Collaborations"] == "yes", "Past Brand Collaborations"] = 1
df.loc[df["Past Brand Collaborations"] == "no", "Past Brand Collaborations"] = 0
```

```python
data.fillna('0',inplace=True)
df.info()
```

```python
df.head()
```

## 4.Feature Engineering`

```python
niche_map = {
    "Digital Marketing": 0.9,
    "Entrepreneurship": 0.9,
    "Technology": 0.85,
    "Career Growth": 0.8,
    "Other": 0.5
}

df["niche_score"] = df["Niche"].map(niche_map).fillna(0.5)
```

```python
business_region_map = {
    # North America
    "United States": "North America",
    "USA": "North America",
    "Canada": "North America",

    # Europe
    "United Kingdom": "Europe",
    "UK": "Europe",
    "Germany": "Europe",
    "France": "Europe",
    "Italy": "Europe",
    "Spain": "Europe",
    "Netherlands": "Europe",
    "Sweden": "Europe",
    "Norway": "Europe",
    "Denmark": "Europe",
    "Finland": "Europe",
    "Switzerland": "Europe",
    "Austria": "Europe",
    "Belgium": "Europe",
    "Ireland": "Europe",
    "Portugal": "Europe",
    "Poland": "Europe",
    "Czech Republic": "Europe",
    "Hungary": "Europe",
    "Romania": "Europe",
    "Greece": "Europe",

    # Oceania
    "Australia": "Oceania",
    "New Zealand": "Oceania",

    # Middle East
    "United Arab Emirates": "Middle East",
    "UAE": "Middle East",
    "Saudi Arabia": "Middle East",
    "Qatar": "Middle East",
    "Kuwait": "Middle East",
    "Bahrain": "Middle East",
    "Oman": "Middle East",
    "Israel": "Middle East",
    "Jordan": "Middle East",

    # South Asia
    "India": "South Asia",
    "Pakistan": "South Asia",
    "Bangladesh": "South Asia",
    "Sri Lanka": "South Asia",
    "Nepal": "South Asia",
    "Bhutan": "South Asia",
    "Maldives": "South Asia",

    # Southeast Asia
    "Singapore": "Southeast Asia",
    "Indonesia": "Southeast Asia",
    "Thailand": "Southeast Asia",
    "Vietnam": "Southeast Asia",
    "Philippines": "Southeast Asia",
    "Malaysia": "Southeast Asia",
    "Cambodia": "Southeast Asia",
    "Laos": "Southeast Asia",
    "Myanmar": "Southeast Asia",
    "Brunei": "Southeast Asia",
    "Timor-Leste": "Southeast Asia",

    # East Asia
    "China": "East Asia",
    "Japan": "East Asia",
    "South Korea": "East Asia",
    "North Korea": "East Asia",
    "Taiwan": "East Asia",
    "Hong Kong": "East Asia",
    "Mongolia": "East Asia",

    # Latin America
    "Brazil": "Latin America",
    "Mexico": "Latin America",
    "Argentina": "Latin America",
    "Chile": "Latin America",
    "Colombia": "Latin America",
    "Peru": "Latin America",
    "Venezuela": "Latin America",
    "Ecuador": "Latin America",
    "Bolivia": "Latin America",
    "Paraguay": "Latin America",
    "Uruguay": "Latin America",
    "Costa Rica": "Latin America",
    "Panama": "Latin America",
    "Guatemala": "Latin America",
    "Honduras": "Latin America",
    "El Salvador": "Latin America",
    "Nicaragua": "Latin America",
    "Dominican Republic": "Latin America",

    # Africa
    "South Africa": "Africa",
    "Nigeria": "Africa",
    "Kenya": "Africa",
    "Ghana": "Africa",
    "Egypt": "Africa",
    "Morocco": "Africa",
    "Tunisia": "Africa",
    "Algeria": "Africa",
    "Ethiopia": "Africa",
    "Tanzania": "Africa",
    "Uganda": "Africa",
    "Rwanda": "Africa",
    "Senegal": "Africa",

    # Eastern Europe & CIS
    "Russia": "Eastern Europe & CIS",
    "Ukraine": "Eastern Europe & CIS",
    "Belarus": "Eastern Europe & CIS",
    "Kazakhstan": "Eastern Europe & CIS",
    "Uzbekistan": "Eastern Europe & CIS",
    "Armenia": "Eastern Europe & CIS",
    "Georgia": "Eastern Europe & CIS",
    "Azerbaijan": "Eastern Europe & CIS",
    "Moldova": "Eastern Europe & CIS",
}


df["Region"] = (df["Location"].map(business_region_map).fillna("Other"))

region_weights = {
    "North America": 1.00,
    "Europe": 0.90,
    "Oceania": 0.85,
    "Middle East": 0.85,
    "East Asia": 0.80,
    "South Asia": 0.75,
    "Southeast Asia": 0.75,
    "Latin America": 0.70,
    "Africa": 0.65,
    "Eastern Europe & CIS": 0.65,
    "Other": 0.40
}

df["location_score"] = df["Region"].map(region_weights)
```

```python
eng_max = df["Engagement Rate (%)"].quantile(0.95)
df["engagement_score"] = df["Engagement Rate (%)"].clip(0, eng_max) / eng_max
```

```python
df["Past Brand Collaborations"] = (
    df["Past Brand Collaborations"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": 1, "no": 0})
    .fillna(0)
    .astype(int)
)

df["brand_collab_score"] = df["Past Brand Collaborations"].astype(float)

df[["Past Brand Collaborations", "brand_collab_score"]].dtypes
```

```python
df["content_frequency_raw"] = (
    df["Average Likes"].fillna(0) /
    df["Follower Count"].replace(0, np.nan)
)

cf_max = df["content_frequency_raw"].quantile(0.95)

df["content_frequency_score"] = (
    df["content_frequency_raw"]
    .clip(0, cf_max)
    / cf_max
)

df["content_frequency_score"] = df["content_frequency_score"].fillna(
    df["content_frequency_score"].median()
)
```

```python
eng_max = df["Engagement Rate (%)"].quantile(0.95)
df["engagement_score"] = df["Engagement Rate (%)"].clip(0, eng_max) / eng_max
```

```python
df["followers_log"] = np.log1p(df["Follower Count"])
df["followers_score"] = (
    df["followers_log"] - df["followers_log"].min()
) / (
    df["followers_log"].max() - df["followers_log"].min()
)
```

```python
df["influencer_score"] = (
    0.30 * df["engagement_score"] +
    0.25 * df["niche_score"] +
    0.15 * df["followers_score"] +
    0.10 * df["brand_collab_score"] +
    0.10 * df["location_score"] +
    0.10 * df["content_frequency_score"]
) * 100

df["influencer_score"] = df["influencer_score"].clip(0, 100)

```

```python
#Encodding

df = pd.get_dummies(
    df,
    columns=["Segment"],
    dtype=int
)

df.head()
```

```python
cols_to_float = [
    "Engagement Rate (%)",
    "engagement_score",
    "content_frequency_raw",
    "influencer_score"
]

df[cols_to_float] = df[cols_to_float].apply(
    pd.to_numeric, errors="coerce"
)
```

```python
df = df.drop(columns=["Engagement Rate (%)","followers_log","content_frequency_raw","brand_collab_score","Past Brand Collaborations","Platform","Region","Niche","Location","Follower Count","Average Likes","Average Comments"])
```

```python
df.fillna('0',inplace=True)
df.info()
```

# REGRESSION MODELS (Forecast Liquidity)

#### Feature Selection & Dataset Splitting

```python
xr = df.drop(["influencer_score"],axis=1)
yr = df["influencer_score"]

from sklearn.model_selection import train_test_split
xrtrain, xrtest, yrtrain, yrtest = train_test_split(xr, yr, test_size=0.2, random_state=42,shuffle=False)
```

```python
print("NaNs in train:\n", pd.isna(xrtrain).sum())
print("NaNs in test:\n", pd.isna(xrtest).sum())
```

#### Model Training, Model Evaluation & Model Training

### 1. LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

lr = LinearRegression()
lr.fit(xrtrain,yrtrain)

pred = lr.predict(xrtest)

print("Linear Regression")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### 2. Ridge

```python
#Perfect model

from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1)
ridge.fit(xrtrain,yrtrain)
pred = ridge.predict(xrtest)

print("Ridge")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### 3. Lasso

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=6)
lasso.fit(xrtrain,yrtrain)
pred = lasso.predict(xrtest)

print("Lasso")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### 4. DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(max_depth=6)
dt.fit(xrtrain,yrtrain)
pred = dt.predict(xrtest)

print("Decision Tree")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor

gb = GradientBoostingRegressor(n_estimators=200,learning_rate=0.05,max_depth=5)
gb.fit(xrtrain,yrtrain)
pred = gb.predict(xrtest)

print("Gradient Boosting")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators=300,max_depth=10,random_state=42)
rf.fit(xrtrain,yrtrain)
pred = rf.predict(xrtest)

print("Random Forest")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

### XGBRegressor

```python
from xgboost import XGBRegressor

xgb = XGBRegressor(n_estimators=120,learning_rate=0.03,max_depth=12,subsample=0.5,colsample_bytree=0.758)
xgb.fit(xrtrain,yrtrain)
pred = xgb.predict(xrtest)

print("XGBoost")
print("MAE:",mean_absolute_error(yrtest,pred))
print("RMSE:",np.sqrt(mean_squared_error(yrtest,pred)))
print("R2:",r2_score(yrtest,pred))
```

# CLASSIFICATION MODELS (Potential Segment)

#### Feature Selection & Dataset Splitting

#### Collaboration Possibilities with Influencer | 🟢 Low Possibility → 0 | 🟡 Medium Possibility → 1 | 🔴 High Possibility → 2 

```python
print(data.columns.tolist())
```

```python
from sklearn.model_selection import train_test_split

features = [
    "engagement_score",
    "niche_score",
    "followers_score",
    "location_score",
    "content_frequency_score",
    "Segment_Macro",
    "influencer_score",
    "Segment_Micro",
    "Segment_Mid-tier"
]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))

scaled_data = scaler.fit_transform(df[features])

scaled_df = pd.DataFrame(
    scaled_data,
    columns=[f"{col}_scaled" for col in features],
    index=df.index
)
scaled_df = scaled_df.dropna()
scaled_df.info()
```

```python
df["collaboration_potential"] = pd.cut(
    scaled_df["influencer_score_scaled"],
    bins=[0, 0.60, 0.75, 1],
    labels=[0, 1, 2],
    include_lowest=True
)
```

```python
df["collaboration_potential"].info()
```

```python
xc = scaled_df

yc = df["collaboration_potential"]

xctrain, xctest, yctrain, yctest = train_test_split(xc,yc,test_size=0.3,random_state=42,shuffle=False)
```

```python
yc.info()
```

#### Model Training, Model Evaluation & Model Training

### 1. Logistic Regression

```python
#Perfect model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix

log = LogisticRegression(max_iter=2000)
log.fit(xctrain,yctrain)

pred = log.predict(xctest)
print("Logistic Regression")
print(classification_report(yctest,pred))
```

### 2. K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(xctrain,yctrain)

pred = knn.predict(xctest)
print("KNN")
print(classification_report(yctest,pred))
```

### 3. Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(xctrain,yctrain)

pred = nb.predict(xctest)
print("Naive Bayes")
print(classification_report(yctest,pred))
```

### 4. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(max_depth=6)
dt.fit(xctrain,yctrain)

pred = dt.predict(xctest)
print("Decision Tree")
print(classification_report(yctest,pred))
```

### 5. Support Vector Machine(SVC)

```python
from sklearn.svm import SVC

svm = SVC(kernel="rbf",C=10,gamma="scale")
svm.fit(xctrain,yctrain)

pred = svm.predict(xctest)
print("SVM")
print(classification_report(yctest,pred))
```

### 6. Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=300,max_depth=10,random_state=42)
rfc.fit(xctrain,yctrain)

pred = rfc.predict(xctest)
print("Random Forest")
print(classification_report(yctest,pred))
```

### 7. Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier(n_estimators=200,learning_rate=0.05,max_depth=5, subsample=0.7)
gbc.fit(xctrain,yctrain)

pred = gbc.predict(xctest)
print("Gradient Boosting")
print(classification_report(yctest,pred))
```

```python
best_classifier = dt
```

## CLASSIFICATION MANUAL TESTING (Risk Regime)

### Collaboration Potential (CP) | 🟢 Low CP → 0 | 🟡 Medium CP → 1 | 🔴 CP → 2 

```python
def interpret_collaboration_potential(prediction):

    if prediction == 2:
        return "🟢 HIGH POTENTIAL: Influencer shows strong engagement and relevance. Recommended for immediate collaboration."

    elif prediction == 1:
        return "🟡 MEDIUM POTENTIAL: Influencer has average performance and relevance. Suitable for selective or trial collaborations with regular performance monitoring."

    else:
        return "🔴 LOW POTENTIAL: Influencer shows weak engagement or low relevance. Not recommended for priority collaboration at this stage."
```

### Test Case 1 – Normal Potential Influencer

```python
sample_1 = xctest.iloc[0].values.reshape(1,-2)

print("Low Risk Influencer collaboration chances:", best_classifier.predict(sample_1))

prediction = best_classifier.predict(sample_1)[0]
print()
print(interpret_collaboration_potential(prediction))
```

### Test Case 2 – Medium Potential Influencer

```python
sample_2 = sample_1.copy()
sample_2 = sample_2 * 1.8
prediction = best_classifier.predict(sample_2)[0]

print("Predicted Influencer collaboration chances:", best_classifier.predict(sample_2))
print()
print(interpret_collaboration_potential(prediction))
```

### Test Case 3 – High Potential Influencer

```python
sample_3 = sample_1.copy()
sample_3 = sample_3 * 4
print("Extreme Risk Influencer collaboration chances:", best_classifier.predict(sample_3))

prediction = best_classifier.predict(sample_3)[0]
print()
print(interpret_collaboration_potential(prediction))
```
